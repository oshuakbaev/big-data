```python
# task 10

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
data = pd.read_csv('C:/Users/lucky/Documents/University/7s semestr/big data/github/big-data/Auto_dataset.csv')

# Гистограмма для average-mileage
plt.figure(figsize=(10,6))
sns.histplot(data['average-mileage'])
plt.title('Histogram of Average Mileage')
plt.show()

# Гистограмма для company
plt.figure(figsize=(10,6))
sns.countplot(y=data['company'])
plt.title('Histogram of Companies')
plt.show()
```
